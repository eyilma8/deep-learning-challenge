## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_drop = application_df.drop(columns=['EIN', 'NAME'])
application_df_drop.head(2)
```

```python
# Determine the number of unique values in each column.
application_cat = application_df_drop.dtypes[application_df_drop.dtypes=="object"].index.tolist()
application_df_drop[application_cat].nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df_drop["APPLICATION_TYPE"].value_counts()
app_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = list(app_counts[app_counts < 500].index)
# Replace in dataframe
for app in application_types_to_replace:
    application_df_drop['APPLICATION_TYPE'] = application_df_drop['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_counts_df = application_df_drop['APPLICATION_TYPE'].value_counts()
app_counts_df
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df_drop['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_counts[class_counts < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_drop['CLASSIFICATION'] = application_df_drop['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df_drop['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_drop = pd.get_dummies(application_df_drop)
application_df_drop.head()
```

```python
# Split our preprocessed data into our features and target arrays
x = application_df_drop.drop(columns=['IS_SUCCESSFUL']).values
y = application_df_drop['IS_SUCCESSFUL'].values
```

```python
y
```

```python
# split into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
x_train[0]
```

```python
x_test[0]
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
print(x_train_scaled[0])
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len(x_train_scaled[0])
layer_1_nodes = 80
layer_2_nodes = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_1_nodes,
                          input_dim=num_input_features,
                          activation='relu')
)


# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_2_nodes,

                          activation='relu')
)


# Output layer
nn.add(
    tf.keras.layers.Dense(units=1,

                          activation='sigmoid')
)

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#  Create a callback
import os
from tensorflow.keras.callbacks import ModelCheckpoint
# Define the checkpoint path and filenames
os.makedirs("checkpoints1/",exist_ok=True)
checkpoint_path = "checkpoints1/weights.{epoch:02d}.hdf5"
# Create a callback which saves the weights for every 5 epochs
cp_callback= ModelCheckpoint(filepath=checkpoint_path,save_weights_only=True, period=5)
```

```python
# Train the model
fit_model = nn.fit(x_train_scaled, y_train, epochs=100,callbacks=[cp_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(x_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("alphabetsoup1h.h5")
```
